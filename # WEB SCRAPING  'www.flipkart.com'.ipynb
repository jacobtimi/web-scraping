```python
import pandas as pd
import numpy as np
```

```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReg
my_url = "https://www.flipkart.com/television-store?otracker=nmenu_sub_TVs%20and%20Appliances_0_Televisions&otracker=nmenu_sub_TVs%20%26%20Appliances_0_Television"
uClient = uReg(my_url)
page_html = uClient.read()
uClient.close()
```

```python
page_soup = soup(page_html, "html.parser")
page_soup
```

```python
containers = page_soup.findAll("div", {"id": "container"})
print(len(containers))
print(soup.prettify(containers[0]))
```

```python
content = containers[0].findAll("div", {"class": "_3YgSsQ"})
len(content)
```

```python
container= content[0]
description = container.findAll("a", {"class": "s1Q9rs"})
description[0].text
```

```python
rating = container.findAll("span", {"class": "_2_R_DZ"})
rating[0].text
```

```python
rating = container.findAll("span", {"class": "_2_R_DZ"})
rating[0].text
```

```python
price = container.findAll("div", {"class": "_30jeq3"})
price[0].text
```

```python
Descriptions = []
Ratings =  []
Price = []
for container in content:
    description = container.findAll("a", {"class": "s1Q9rs"})
    descript = description[0].text
    Descriptions.append(descript) 
    
    rating = container.findAll("span", {"class": "_2_R_DZ"})
    RATE = rating[0].text
    Ratings.append(RATE)
    
    price = container.findAll("div", {"class": "_30jeq3"})
    price_product = price[0].text
    Price.append(price_product)
    
    print("description:"+ descript)
    print("rating:"+ RATE)
    print("price:"+ price_product)
```

```python
TV_product = pd.DataFrame({"Product  Name & Description":Descriptions, "Rating":Ratings, "Price": Price})
TV_product.to_csv("tv_products.csv")
tv = pd.read_csv("tv_products.csv")
tv
```
